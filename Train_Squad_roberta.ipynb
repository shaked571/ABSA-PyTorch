```python
data = pd.read_csv("/content/nps_data/res_test_my_aspects.csv")
```

```python
questions_template = ["What is good in Amazon?", 
                      "What is good in Walmart?"]

```

```python
cols = [d for d in data]
res_amzn_min = []
res_walmart_min = []
for t in data.text.to_list():
    amz_input = {
    'question': 'What is good in Amazon?',
    'context': t
    }  
    res = nlp(amz_input)
    if res['score'] > 0.001:
        res_amzn_min.append(res['answer'])
    else:
      res_amzn_min.append("NA")

    wmt_input = {
    'question': 'What is good in Walmart?',
    'context': t
    }  
    res = nlp(wmt_input)
    if res['score'] > 0.001:
        res_walmart_min.append(res['answer'])
    else:
      res_walmart_min.append("NA")




```

```python
data['walmart_min_0.001'] = res_walmart_min
data['amzn_min_0.001'] = res_amzn_min
```
